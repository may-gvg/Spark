```
rdd = sc.parallelize([
    "this is positive message",
    "this is negative comment",
    "very positive and happy"
])
```

```
#ładujemy, tworzymy  dane do rdd
```

```
rdd.count()
```

```
rdd.take(10)
```

```
# pokazujemy pobrane dane
```

```
def simple_sentiment(text):
    if "positive" in text:
        return "positive"
    else:
        return "negative"

def fun(a):
    return (simple_sentiment(a), 1)

rdd.map(fun).take(10)
```

```
# sprawdzamy ile mamy pozytywnych a ile negatywnych opinii, dzięki funkcji tworzymy mapowanie czyli na zasadzie słownika przyporządkować każdemu elementowi naszej kolekcji jakąś wartość
```

```
rdd.map(fun).reduceByKey(lambda a,b: a+b).take(10)
```

```
# dzięki redukcji, tworzymy nowy obiekt rdd będący sumą rodzaju opini, który jest wyświtlony poprzez take
```
